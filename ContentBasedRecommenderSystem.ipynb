```python
import ast

string_list_of_dicts = '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

# Convert the string to a list of dictionaries
list_of_dicts = ast.literal_eval(string_list_of_dicts)

print(type(list_of_dicts))
print(list_of_dicts)
```

```python
# we need to bring all the name values of above list of dicts

def convert(obj):
  L = []
  for i in ast.literal_eval(obj) :
    L.append(i["name"])
  return L
```

```python
# we just need top 3 casting
def convert_cast(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj) :
    if counter != 3:
      L.append(i["name"])
      counter += 1
    else :
      break
  return L
```

```python
movies["cast"] = movies["cast"].apply(convert_cast)
```

```python
# we only need director

def convert_crew(obj):
  L = []
  for i in ast.literal_eval(obj) :
    if i["job"] == "Director":
      L.append(i["name"])
      break
  return L
```

```python
movies["crew"] = movies["crew"].apply(convert_crew)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

# We are using stemming because our aim in this project is to find the similarity of the movies based on the tags column.
# Here, the root word replaces all its derived forms, so that words like "dancing", "danced", "dance" are all represented by the same root "danc".
# This ensures that different forms of the same concept are counted together when creating vectors.
# As a result, the similarity between movies reflects the presence of the same ideas or themes in their tags, rather than being affected by variations in word forms.
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
movies_new['tags'] = movies_new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = "english") ## it removes stopwords in english
```

The max_features parameter in CountVectorizer limits the size of the vocabulary to the top N most frequent words in the dataset. By default, CountVectorizer includes all unique words it finds. However, in large text datasets, this can create a huge vocabulary (tens of thousands of words), which increases memory usage and may add noise. Setting max_features ensures that only the most common words (based on term frequency across all documents) are kept, and all rarer words are ignored. For example, if max_features=5000, only the 5,000 most frequent words will be used to create the document-term matrix, making the model more efficient and reducing overfitting to rare words.

1. What CountVectorizer Does

It tokenizes the text (splits into words).

Builds a vocabulary (list of all unique words across the dataset).

Transforms each document (row of text) into a vector of word counts.

🔹 2. Vocabulary Order

The vocabulary is not sorted by frequency.

By default, it is stored in alphabetical (lexicographic) order of the tokens.

Example:
If the dataset has the words "thriller", "action", "movie", "romantic",
then the vocabulary will be:

['action', 'movie', 'romantic', 'thriller']

🔹 3. Transforming Rows

Each row of text is turned into a vector of counts, where:

Rows = documents (movie tags)

Columns = words from the vocabulary

Values = how many times that word appears in the document

🔹 4. Example

Suppose movies_new['tags'] has:

Row 1: "action action thriller movie"
Row 2: "romantic action movie"


Vocabulary (alphabetical order):
['action', 'movie', 'romantic', 'thriller']

Transformation:

Row 1 → [2, 1, 0, 1]
(2× action, 1× movie, 0× romantic, 1× thriller)

Row 2 → [1, 1, 1, 0]
(1× action, 1× movie, 1× romantic, 0× thriller)

Final output (.toarray() gives a dense NumPy array):

[[2 1 0 1]
 [1 1 1 0]]

🔹 5. Key Takeaway

CountVectorizer does not sort words by frequency.

Words are stored in alphabetical order by default.

The .toarray() result is a document-term matrix, where:

Each row = one document (tags of a movie).

Each column = one vocabulary word.

Each value = word count in that document.\

```python
# use cosine similarity for similarity check

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[1])), reverse = True, key = lambda x : x[1])[1:6]

# this is to give index(using enumerate) to the each value, to get to know the index value when sorted
# and sorted() with key parameter, which gives custom col to sort on
```

```python
movies_new[movies_new["title"] == "Batman Begins"].index
```

```python
def recommend(movie):
  movie_index = movies_new[movies_new["title"] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x : x[1])[1:6]

  for i in movies_list :
    print(movies_new.iloc[i[0]]["title"])
```

```python
recommend("Avatar")
```

```python
recommend('Batman Begins')
```
